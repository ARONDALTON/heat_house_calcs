```python
import pandas as pd
import numpy as np
import pint
```

```python
ur = pint.UnitRegistry()
dir(ur )
#3 * ur.m + 4 * ur.cm
```

```python
air_exchange_per_hr=[.25,.5,.75,1,2,3,4,5]
cost_of_electrcity=.129 #ur.$#$/kWh
outside_to_inside_temp_difference=[5,10,15,20]  #deg C
sq_ft_of_space=1000 *ur.ft**2#sqr-ft
wall_height=10 *ur.ft#ft
volume_of_room=sq_ft_of_space*wall_height
print('volume_of ',sq_ft_of_space,' room is: ',volume_of_room)#, ' cu-ft')
```

```python
density_of_air= .036 *ur.kg/ur.ft**3#Kg/cu.ft
specific_heat_of_air=1.006 *1000*ur.J /ur.kg/ur.C#kJ/kgC

wattage_of_heater=1000 *ur.W# W

mass_of_air=density_of_air*volume_of_room
print('mass_of_air is: ', mass_of_air)#, ' kg')
```

```python
def convert_to_cost(mass_of_air,specific_heat_of_air,density_of_air,volume_of_room,wattage_of_heater,outside_to_inside_temp_difference):
    volume_of_room=sq_ft_of_space*wall_height
    print(volume_of_room)
    mass_of_air=density_of_air*volume_of_room
    print(mass_of_air)
    energy_needed=specific_heat_of_air*outside_to_inside_temp_difference*ur.C*mass_of_air
    print('energy_needed is: ',energy_needed)#,' kJ or kW.sec??')
    time_to_heat=energy_needed*1000/wattage_of_heater
    print('time_to_heat is ',time_to_heat, ' seconds, or ',wattage_of_heater,' W heater is on for ',time_to_heat/(60*60))#, 'hr')
    wattage_of_heater*time_to_heat/(60*60)/1000*cost_of_electrcity
    cost=lambda energy_needed:(energy_needed/(60*60))* (cost_of_electrcity)
    
    return (cost(energy_needed.to_reduced_units())*24),'\n',(cost(energy_needed.to_reduced_units())*24*30)
```

```python

[convert_to_cost(mass_of_air,specific_heat_of_air,density_of_air,volume_of_room,wattage_of_heater,i)
   for i in outside_to_inside_temp_difference ]
```
